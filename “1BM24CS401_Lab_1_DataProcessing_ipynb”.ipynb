```python
import pandas as pd

# Load CSV file from Colab environment
df1 = pd.read_csv("/content/adult.csv")

# Display first 5 rows
df1.head()
```

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler, StandardScaler


```

```python
df.isna().sum()
```

```python
#(b) Handling Categorical Data
# Identify categorical columns
cat_cols_diabetes = df.select_dtypes(include='object').columns
cat_cols_diabetes
```

```python
# Label Encoding
le = LabelEncoder()

for col in cat_cols_diabetes:
    df[col] = le.fit_transform(df[col])
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
df
```

```python
#DATA TRANSFORMATION  Diabetes Dataset
minmax = MinMaxScaler()
df_minmax = minmax.fit_transform(df)
df_minmax
```

```python
standard = StandardScaler()
df_standard = standard.fit_transform(df)
df_standard

```

```python
# Replace '?' with NaN
df1.replace('?', np.nan, inplace=True)

# Check missing values
df1.isna().sum()
```

```python
for col in df1.columns:
    if df1[col].isna().sum() > 0:
        df1[col].fillna(df1[col].mode()[0], inplace=True)
```

```python
# Identify categorical columns
cat_cols_adult = df1.select_dtypes(include='object').columns
cat_cols_adult
```

```python
# Label Encoding
for col in cat_cols_adult:
    df1[col] = le.fit_transform(df1[col])
```

```python
Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1

df1 = df1[~((df1 < (Q1 - 1.5 * IQR)) | (df1 > (Q3 + 1.5 * IQR))).any(axis=1)]
df1
```

```python
#DATA TRANSFORMATION  Adult Income Dataset
#(a) Min-Max Scaling
df1_minmax = minmax.fit_transform(df1)
df1_minmax
```
