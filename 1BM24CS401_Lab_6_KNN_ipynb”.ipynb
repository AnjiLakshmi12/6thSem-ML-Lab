```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the dataset
iris_df = pd.read_csv('/content/iris (1).csv')
print("Iris DataFrame head:")
print(iris_df.head())

# Separate features (X_iris) and target (y_iris)
X_iris = iris_df.iloc[:, :-1] # All columns except the last one as features
y_iris = iris_df.iloc[:, -1]  # The last column as target

print("\nFeatures (X_iris) head:")
print(X_iris.head())
print("\nTarget (y_iris) head:")
print(y_iris.head())

# Split the data into training and testing sets (80% train, 20% test)
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(X_iris, y_iris, test_size=0.2, random_state=42)

print(f"\nShape of X_train_iris: {X_train_iris.shape}")
print(f"Shape of X_test_iris: {X_test_iris.shape}")
print(f"Shape of y_train_iris: {y_train_iris.shape}")
print(f"Shape of y_test_iris: {y_test_iris.shape}")
```

## KNN for Iris Dataset

### Subtask:
Build a KNN classifier for the Iris dataset. Choose an appropriate 'k' value, train the model, predict on the test set, and then display the accuracy score, plot the confusion matrix, and print the classification report.


**Reasoning**:
The subtask requires building a KNN classifier, training it, making predictions, and then evaluating its performance using accuracy, confusion matrix, and a classification report. The provided instructions detail each step to achieve this, so I will combine them into a single code block.



```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Instantiate a KNeighborsClassifier with n_neighbors = 5
knn_iris = KNeighborsClassifier(n_neighbors=5)
print("KNN Classifier instantiated with n_neighbors=5.")

# Train the model
knn_iris.fit(X_train_iris, y_train_iris)
print("KNN model trained successfully.")

# Make predictions on the test data
y_pred_iris = knn_iris.predict(X_test_iris)
print("Predictions made on the test set.")

# Calculate and print the accuracy score
accuracy = accuracy_score(y_test_iris, y_pred_iris)
print(f"\nAccuracy Score: {accuracy:.4f}")

# Generate the confusion matrix
conf_matrix = confusion_matrix(y_test_iris, y_pred_iris)
print("\nConfusion Matrix generated.")

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=knn_iris.classes_, yticklabels=knn_iris.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for Iris KNN Classifier')
plt.show()

# Print the classification report
print("\nClassification Report:")
print(classification_report(y_test_iris, y_pred_iris))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# 1. Load the 'diabetes.csv' file into a pandas DataFrame
diabetes_df = pd.read_csv('/content/diabetes.csv')
print("Diabetes DataFrame head:")
print(diabetes_df.head())

# 2. Separate features (X_diabetes) and target (y_diabetes)
X_diabetes = diabetes_df.iloc[:, :-1] # All columns except the last one as features
y_diabetes = diabetes_df.iloc[:, -1]  # The last column as target

print("\nFeatures (X_diabetes) head:")
print(X_diabetes.head())
print("\nTarget (y_diabetes) head:")
print(y_diabetes.head())

# 3. Split the data into training and testing sets (80% train, 20% test)
X_train_diabetes, X_test_diabetes, y_train_diabetes, y_test_diabetes = train_test_split(X_diabetes, y_diabetes, test_size=0.2, random_state=42)

# 4. Print the shapes of the resulting training and testing sets
print(f"\nShape of X_train_diabetes: {X_train_diabetes.shape}")
print(f"Shape of X_test_diabetes: {X_test_diabetes.shape}")
print(f"Shape of y_train_diabetes: {y_train_diabetes.shape}")
print(f"Shape of y_test_diabetes: {y_test_diabetes.shape}")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Instantiate StandardScaler
scaler = StandardScaler()
print("StandardScaler instantiated.")

# 2. Fit the scaler on the training data and transform both training and test data
X_train_diabetes_scaled = scaler.fit_transform(X_train_diabetes)
X_test_diabetes_scaled = scaler.transform(X_test_diabetes)
print("Features scaled for both training and test sets.")

# 3. Instantiate a KNeighborsClassifier with n_neighbors = 5
knn_diabetes = KNeighborsClassifier(n_neighbors=5)
print("KNN Classifier instantiated with n_neighbors=5.")

# 4. Train the model using the scaled training data
knn_diabetes.fit(X_train_diabetes_scaled, y_train_diabetes)
print("KNN model trained successfully on scaled data.")

# 5. Make predictions on the scaled test data
y_pred_diabetes = knn_diabetes.predict(X_test_diabetes_scaled)
print("Predictions made on the scaled test set.")

# 6. Calculate and print the accuracy score
accuracy_diabetes = accuracy_score(y_test_diabetes, y_pred_diabetes)
print(f"\nAccuracy Score (Diabetes): {accuracy_diabetes:.4f}")

# 7. Generate the confusion matrix
conf_matrix_diabetes = confusion_matrix(y_test_diabetes, y_pred_diabetes)
print("\nConfusion Matrix generated for Diabetes dataset.")

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_diabetes, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for Diabetes KNN Classifier (Scaled Data)')
plt.show()

# 8. Print the classification report
print("\nClassification Report (Diabetes):")
print(classification_report(y_test_diabetes, y_pred_diabetes))
```

## KNN for Heart Dataset with Optimal K

### Subtask:
Build a KNN classifier for the Heart dataset. Iterate through different 'k' values to find the one that yields the maximum classification score. Train the model with this optimal 'k', then plot the confusion matrix and print the classification report for the best model.


**Reasoning**:
The subtask requires building a KNN classifier for the Heart dataset, finding the optimal 'k' value by iterating through a range of 'k's, training the model with the best 'k', and then evaluating its performance with a confusion matrix and classification report. The provided instructions detail each step to achieve this, so I will combine them into a single code block.



```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Initialize an empty list to store accuracy scores for different 'k' values
#    and variables to store the best 'k' and its corresponding best accuracy.
accuracies = []
best_k = 0
best_accuracy = 0

print("Iterating through k values to find the optimal k...")

# 2. Iterate through a range of 'k' values (e.g., from 1 to 20)
for k in range(1, 21): # K values from 1 to 20
    # 3a. Instantiate a KNeighborsClassifier with the current 'k' value.
    knn_heart = KNeighborsClassifier(n_neighbors=k)

    # 3b. Train the model using X_train_heart and y_train_heart.
    knn_heart.fit(X_train_heart, y_train_heart)

    # 3c. Make predictions on X_test_heart.
    y_pred_heart = knn_heart.predict(X_test_heart)

    # 3d. Calculate the accuracy score and append it to the accuracy list.
    current_accuracy = accuracy_score(y_test_heart, y_pred_heart)
    accuracies.append(current_accuracy)

    # 3e. Update the best_k and best_accuracy if the current accuracy is higher.
    if current_accuracy > best_accuracy:
        best_accuracy = current_accuracy
        best_k = k

# 4. Print the best_k and best_accuracy found.
print(f"\nOptimal k found: {best_k}")
print(f"Best accuracy with optimal k: {best_accuracy:.4f}")

# 5. Instantiate a new KNeighborsClassifier using the best_k found.
final_knn_heart = KNeighborsClassifier(n_neighbors=best_k)

# 6. Train this final model using X_train_heart and y_train_heart.
final_knn_heart.fit(X_train_heart, y_train_heart)
print(f"\nFinal KNN model trained with optimal k={best_k}.")

# 7. Make predictions on X_test_heart using the final model.
y_pred_final_heart = final_knn_heart.predict(X_test_heart)
print("Predictions made using the final model.")

# 8. Generate and plot the confusion matrix
conf_matrix_heart = confusion_matrix(y_test_heart, y_pred_final_heart)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_heart, annot=True, fmt='d', cmap='Blues',
            xticklabels=final_knn_heart.classes_, yticklabels=final_knn_heart.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title(f'Confusion Matrix for Heart KNN Classifier (Optimal k={best_k})')
plt.show()

# 9. Print the classification report
print("\nClassification Report for Heart Dataset (Optimal k):")
print(classification_report(y_test_heart, y_pred_final_heart))
```
