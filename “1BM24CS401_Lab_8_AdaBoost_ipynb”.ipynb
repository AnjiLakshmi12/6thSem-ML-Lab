```python
import pandas as pd
```

```python
df = pd.read_csv('/content/income.csv')
display(df.head())
```

### 1. Data Preparation
First, we'll separate the features (X) from the target variable (y), which is 'income_level'. Then, we'll split the data into training and testing sets.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Separate features (X) and target (y)
X = df.drop('income_level', axis=1)
y = df['income_level']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training set shape: {X_train.shape}")
print(f"Testing set shape: {X_test.shape}")
```

### 2. Build AdaBoost Classifier (n_estimators=10)
Now, let's build an AdaBoost classifier with `n_estimators=10` and evaluate its prediction score.

```python
# Initialize the base estimator (Decision Tree Classifier is common for AdaBoost)
# Using a small max_depth to prevent overfitting of individual trees
b_estimator = DecisionTreeClassifier(max_depth=1, random_state=42)

# Initialize AdaBoost classifier with n_estimators=10
adaboost_clf_initial = AdaBoostClassifier(
    estimator=b_estimator,
    n_estimators=10,
    random_state=42
)

# Train the model
adaboost_clf_initial.fit(X_train, y_train)

# Make predictions on the test set
y_pred_initial = adaboost_clf_initial.predict(X_test)

# Measure the prediction score (accuracy)
initial_accuracy = accuracy_score(y_test, y_pred_initial)
print(f"Prediction score with n_estimators=10: {initial_accuracy:.4f}")
```

### 3. Fine-tune `n_estimators`
Next, we'll fine-tune the `n_estimators` parameter (number of boosting rounds/trees) to identify the best possible score.

```python
best_score = 0
best_n_estimators = 0

# Define a range of n_estimators to test
estimators_range = range(1, 201, 10) # Test from 1 to 200 with steps of 10

print("Fine-tuning AdaBoost Classifier...")
for n_est in estimators_range:
    adaboost_clf = AdaBoostClassifier(
        estimator=b_estimator,
        n_estimators=n_est,
        random_state=42
    )
    adaboost_clf.fit(X_train, y_train)
    y_pred = adaboost_clf.predict(X_test)
    score = accuracy_score(y_test, y_pred)

    if score > best_score:
        best_score = score
        best_n_estimators = n_est
    # print(f"n_estimators: {n_est}, Score: {score:.4f}") # Uncomment to see all scores

print(f"\nBest prediction score achieved: {best_score:.4f}")
print(f"using n_estimators (number of trees): {best_n_estimators}")
```
